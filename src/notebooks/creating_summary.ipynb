```python
file_path = '../../data/PatientsData/p001_20260502_113117/20260502_113117_01_Tutorial_events.csv'
```

```python
with open(file_path, "r", encoding="utf-8") as f:
    # read all lines from the file
    lines = f.readlines()

# extract the first line (header) and split it into columns
columns_line = lines[0]
columns = columns_line.strip().split(",")
lines.pop(0)  # remove the header line from the list of lines
print(columns)

# separate lines based on the section break
pattern = r"^--- SECTION BREAK:"
selected_lines = []

for line in lines:
    if re.match(pattern, line):
        #df = pd.DataFrame(selected_lines, columns=columns)
        break
    selected_lines.append(line)

print(selected_lines)
```

```python
import pandas as pd

# Read the CSV
#df = pd.read_csv("input.csv", header=None)
df = tutorial_events.copy()  # Using the tutorial_events DataFrame for demonstration
# Find rows containing the separator
separator = "--- SECTION BREAK"

# Create groups based on separator occurrences
groups = (df[0] == separator).cumsum()

# Remove separator rows
df_clean = df[df[0] != separator]

# Split and save
for i, (_, group) in enumerate(df_clean.groupby(groups[df_clean.index]), start=1):
    output_file = f"output_{i}.csv"
    group.to_csv(output_file, index=False, header=False)

    print(f"Created: {output_file}")
```
